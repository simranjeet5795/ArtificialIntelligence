# Reshaping

```python
orig_tensor = torch.tensor([[[1,2,3],[4,7,9],[7,8,9]]])
print(orig_tensor.shape)
print(orig_tensor)
reshaped_tensor = orig_tensor.reshape(3,3) #unalble to understand the differenc
reshaped2 = orig_tensor.reshape(1,9)
view= orig_tensor.view(3,1,3)
print(f"Reshaped Tensor Shape: {reshaped_tensor.shape}")
print(f"Reshaped Tensor:\n{reshaped_tensor}\n")
print(f"Reshaped 2 Shape: {reshaped2.shape}") 
print(f"Reshaped 2:\n{reshaped2}\n")
print(f"View Shape: {view.shape}") 
print(f"view:\n{view}\n")


```

# Difference in permute() and reshape() and view()

```python
permuted_tensor = orig_tensor.permute(2,1,0)
print(permuted_tensor)
```
